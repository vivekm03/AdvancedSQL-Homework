# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite", echo=False)
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
### Precipitation Analysis
#For Honolulu station (Also including precipitation by a specific station, since the data has more than one measurement
#per date, so the next analysis for all stations did not seem intutive)
# Design a query to retrieve the last 12 months of precipitation data and plot the results
# Calculate the date 1 year ago from the last data point in the database where station is Honolulu

hono_lastdt = session.query(func.max(Measurement.date))\
    .filter(Measurement.station == (session.query(Station.station).filter(Station.name.ilike("%honolulu%")))).scalar()
hono_yearago=datetime.strptime(hono_lastdt, '%Y-%m-%d').date() - dt.timedelta(days=365)
print(hono_yearago)

# Perform a query to retrieve the data and precipitation scores
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
data_analysis = pd.read_sql(session.query(Measurement)\
    .filter(Measurement.station == (session.query(Station.station).filter(Station.name.ilike("%honolulu%"))))\
    .filter(Measurement.date>=hono_yearago).statement,session.bind)
data_analysis.set_index(keys="date", inplace=True)
data_analysis.sort_values(by="date", inplace=True)

# Use Pandas Plotting with Matplotlib to plot the data
data_analysis["prcp"].plot(figsize=(12,10))
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
plt.title("Precipitation for Honolulu Station for a year")
plt.savefig('prcp.png', bbox_inches='tight')
```

```python
### Precipitation Analysis
#For All stations
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database where station is Honolulu

all_lastdt = session.query(func.max(Measurement.date)).scalar()
all_yearago=datetime.strptime(all_lastdt, '%Y-%m-%d').date() - dt.timedelta(days=365)
print(all_yearago)

# Perform a query to retrieve the data and precipitation scores
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
data_analysis_all = pd.read_sql(session.query(Measurement).filter(Measurement.prcp != "")\
    .filter(Measurement.date>=all_yearago).statement,session.bind)
data_analysis_all.sort_values(by="date", inplace=True)
data_analysis_all["precipitation"]=data_analysis_all["prcp"].astype(float)
data_analysis_fin=data_analysis_all.set_index('date').drop('id', axis=1)
data_analysis_all.sort_values(by="date", inplace=True)

# Use Pandas Plotting with Matplotlib to plot the data
print(data_analysis_fin.head())
fig, ax = plt.subplots()
data_analysis_fin["precipitation"].plot(figsize=(12,10))
plt.savefig('prcp.png', bbox_inches='tight')
#data_analysis_fin.to_csv("chk_data_analysis_fin.csv")
```

```python
### Station Analysis
# Design a query to show how many stations are available in this dataset?
numst=session.query(Measurement.station).distinct().count()
print(f"Total number of stations: {numst}")
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
q_active=session.query(Measurement.station,func.count(Measurement.station))\
    .group_by(Measurement.station)\
    .order_by(func.count(Measurement.station).desc()).all()
print("Most Active Stations:")
q_active
```

```python
### Station Analysis
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs))\
    .filter(Measurement.station==q_active[0][0])\
    .group_by(Measurement.station)\
    .all()

```

```python
### Station Analysis
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database where station is most active
maxst_lastdt = session.query(func.max(Measurement.date))\
    .filter(Measurement.station==q_active[0][0]).scalar()
maxst_yearago=datetime.strptime(maxst_lastdt, '%Y-%m-%d').date() - dt.timedelta(days=365)
print(maxst_yearago)

# Perform a query to retrieve the data and precipitation scores
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
data_analysis_temp = pd.read_sql(session.query(Measurement.date, Measurement.tobs)\
    .filter(Measurement.station==q_active[0][0])\
    .filter(Measurement.date>=maxst_yearago)\
    .statement,session.bind)
data_analysis_temp.set_index(keys="date", inplace=True)
data_analysis_temp.sort_values(by="date", inplace=True)

#Draw figure
data_analysis_temp["tobs"].plot.hist(bins=12, alpha=0.5)
plt.savefig('temp.png', bbox_inches='tight')
```

```python
### Temperature Analysis
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d     
    Returns:
        TMIN, TAVE, and TMAX
    """  
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
### Temperature Analysis
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip_temp=calc_temps('2015-01-15', '2015-01-21')

#convert tuples to lists using list comprehension
tt_list = [list(i) for i in trip_temp]
tt_list
```

```python
### Temperature Analysis
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
# Use Pandas Plotting with Matplotlib to plot the data

fig, ax = plt.subplots(figsize=(4,10))
ax.bar(x="", height=tt_list[0][1],yerr=tt_list[0][2]-tt_list[0][0],color="darksalmon",alpha=0.6)
ax.set_ylabel("Temp(F)")
ax.set_title("Trip Avg Temp")
fig.tight_layout()
plt.show()

```

```python
### Other Recommended Analysis
# Calculate the rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
session.query(Station.station, Station.name, Station.latitude, Station.longitude, Station.elevation, Measurement.prcp)\
    .filter(Measurement.station == Station.station)\
    .filter(Measurement.date >= '2015-01-15')\
    .filter(Measurement.date <= '2015-01-21')\
    .order_by(Measurement.prcp.desc())\
    .all()
```

```python
### Other Recommended Analysis
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
### Other Recommended Analysis
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
stdt="2015-01-15"
endt="2015-01-21"
# Use the start and end date to create a range of dates
daterange= pd.date_range(stdt, endt).tolist()
# Stip off the year and save a list of %m-%d strings
md_list=[date.strftime("%m-%d") for date in daterange]
# Loop through the list of %m-%d strings and calculate the normals for each date
ls_tmin=[]
ls_tavg=[]
ls_tmax=[]
for date in md_list:
    ls_tmin.append(daily_normals(date)[0][0])
    ls_tavg.append(daily_normals(date)[0][1])
    ls_tmax.append(daily_normals(date)[0][2])
```

```python
### Other Recommended Analysis
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
d = {'Date': daterange, 'Tmin': ls_tmin, 'Tavg': ls_tavg, 'Tmax': ls_tmax}
df = pd.DataFrame(data=d).set_index("Date")
df
```

```python
### Other Recommended Analysis
# Plot the daily normals as an area plot with `stacked=False`
df.plot.area(stacked=False)
```
